```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

## Results 30 Iterations Naive Bayes DDM Agraw1 Abrupt - 10k

```python
df_concat = pd.concat(map(pd.read_csv, glob.glob('abrupt_agraw1/bayes.NaiveBayes_DDM_10000_*_AGRAW1_Abrupt.csv')))
#print(np.mean(df_concat['classifications correct (percent)']))
df_concat.describe()
```

```python
median = []
for file in glob.glob('abrupt_agraw1/bayes.NaiveBayes_DDM_10000_*_AGRAW1_Abrupt.csv'):
    df = pd.read_csv(file)
    #print(file)
    median.append(df.iloc[-1]['classifications correct (percent)'])
    #print(df.iloc[-1]['classifications correct (percent)'])

#print(len(median))
print(f'avg of all last execution: {np.mean(median)}')
```
